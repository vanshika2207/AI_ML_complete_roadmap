```python
def check_even_list(num_list):
    for number in num_list:
        if number%2==0:
            return True 
        else:
            pass
            # return False wrong
    return False        
```

```python
check_even_list([1,3,5])
```

```python
check_even_list([2,4,5])
```

```python
check_even_list([3,4])
```

```python
check_even_list([1,5,7])
```

```python
def get_even_list(num_list):
    even_list=[]
    for number in num_list:
        if number%2==0:
            even_list.append(number)
        else:
            pass
            # return False wrong
    return even_list
```

```python
get_even_list([4,1,2,3,3,6,7,8,0])
```
