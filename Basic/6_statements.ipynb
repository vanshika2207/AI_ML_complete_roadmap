Control Flow - we only want certain code to execute when a particular condition has been met . To control this flow of logic we use some keywords .
- if
- elif
- else

control flow syntax use of colons and indentations.

<code>
if some_condition:
  execute some code
elif some_other_condition:
  do something different
else:
  do something else
</code>  
       

```python
if True:
    print("IT's true")
```

while loops

while loops will continue to execute a block of code while some condition remains True

while some_boolean_condition:
   do something


```python
x=0
while x<5:
    print(f'the current value of x is {x}')
    x+=1
else:
    print('X is not less than 5')    
  
```

### break continue and pass 
break-> breaks out of the current closest encoding loop.
continue-> goes to the top of the closest encoding loop
pass-> does nothing at all

```python
x=[1,2,3]
for item in x:
    pass
print('end of my script')
```

```python
mystring='vanshi'
for letter in mystring:
    if letter=='a':
        continue
    print(letter)
```

```python
mystring='vanshi'
for letter in mystring:
    if letter=='s':
        break
    print(letter)
```

```python
x=0
while x<5:
    print(x)
    if x==2:
        break
    x+=1
```

#### LIST COMPREHENSION
- list comprehensions are a unique way of quickly creating a list with python.
alternative to for+append

```python
mystring='hello'
mylist=[]
for letter in mystring:
    mylist.append(letter)
```

```python
mylist
```

```python
mylist=[letter for letter in mystring]
```

```python
mylist
```

```python
mylist =[x for x in mylist]
```

```python
mylist
```

```python
mylist=[x for x in range(0,11)]
mylist
```

```python
mylist=[x*x for x in range(0,11)]
mylist
```

```python
mylist=[x for x in range(0,11) if x%2==0]
mylist
```

```python
celcius=[0,10,20,34.5]
fahrenheit=[((9/5)*temp +32) for temp in celcius]
```
